```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load a local embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')  # 384-dim vectors, fast and good

# Example multiline text

# Step 1: Split into individual sentences or lines
sentences = [s.strip() for s in text.strip().split('\n') if s.strip()]

# Step 2: Generate embeddings
embeddings = model.encode(sentences)

# Step 3: Semantic chunking based on similarity threshold
```

```python
def semantic_chunking(sentences, embeddings, threshold=0.3):
    chunks = []
    current_chunk = [sentences[0]]

    for i in range(1, len(sentences)):
        sim = cosine_similarity([embeddings[i]], [embeddings[i - 1]])[0][0]
        if sim > threshold:
            current_chunk.append(sentences[i])
        else:
            chunks.append(" ".join(current_chunk))
            current_chunk = [sentences[i]]

    chunks.append(" ".join(current_chunk))
    return chunks

# Run chunking
chunks = semantic_chunking(sentences, embeddings)

# Display results
for i, chunk in enumerate(chunks):
    print(f"\n🧩 Chunk {i + 1}:\n{chunk}")
```

```python
from sentence_transformers import CrossEncoder

# Load the reranker model
reranker = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')

# Step 3: Simple reranker function
def rerank(query, documents, top_k=3):
    """
    Rerank documents based on relevance to query

    Args:
        query (str): The search query
        documents (list): List of text documents or Document objects
        top_k (int): Number of top results to return

    Returns:
        list: Top reranked documents with scores
    """

    # Handle both strings and Document objects
    if hasattr(documents[0], 'page_content'):
        # LangChain Document objects
        texts = [doc.page_content for doc in documents]
    else:
        # Plain strings
        texts = documents

    # Create query-document pairs
    pairs = [[query, text] for text in texts]

    # Get relevance scores
    scores = reranker.predict(pairs)

    # Combine documents with scores
    results = []
    for i, (doc, score) in enumerate(zip(documents, scores)):
        results.append({
            'document': doc,
            'score': float(score),
            'rank': i + 1
        })

    # Sort by score (highest first) and return top_k
    sorted_results = sorted(results, key=lambda x: x['score'], reverse=True)
    return sorted_results[:top_k]
```

```python


def create_medical_simple_prompt(query: str, retrieved_docs) -> str:
    """
    Version spécialisée pour la médecine éthique avec disclaimers
    """

    # Conversion si documents LangChain
    if retrieved_docs and hasattr(retrieved_docs[0], 'page_content'):
        docs_content = []
        for i, doc in enumerate(retrieved_docs):
            docs_content.append({
                'content': doc.page_content,
                'id': doc.metadata.get('id', i+1),
                'title': doc.metadata.get('title', f'Document {i+1}')
            })
    else:
        docs_content = [{'content': doc['content'],
                        'id': doc.get('source_id', i+1),
                        'title': doc.get('title', f'Document {i+1}')}
                       for i, doc in enumerate(retrieved_docs)]

    # Construction du contexte médical
    context = ""
    for i, doc in enumerate(docs_content, 1):
        context += f"Source {i} - {doc['title']} (ID: {doc['id']}):\n{doc['content']}\n\n"

    # Prompt médical avec sécurité
    prompt = f"""Tu es un assistant en éthique médicale. Réponds en utilisant uniquement les sources fournies.

SOURCES MÉDICALES:
{context}

QUESTION: {query}

RÈGLES:
- Réponds uniquement avec les informations des sources
- Cite tes sources (ex: selon Source 1)
- Ne donne jamais de conseils médicaux directs
- Ajoute un disclaimer médical

RÉPONSE:"""

    return prompt

async def generate_simple_response(query: str, retrieved_docs, llm) -> str:
    """
    Génère une réponse simple avec le prompt et le LLM

    Args:
        query: Question
        retrieved_docs: Documents récupérés
        llm: Instance ChatMistralAI configurée

    Returns:
        str: Réponse générée
    """

    # Création du prompt
    prompt = create_medical_simple_prompt(query, retrieved_docs)

    # Génération avec le LLM
    from langchain.schema import HumanMessage

    try:
        response = await asyncio.to_thread(llm.invoke, [HumanMessage(content=prompt)])
        answer = response.content if hasattr(response, 'content') else str(response)

        # Ajout disclaimer si pas présent
        if 'consulter' not in answer.lower() and 'médecin' not in answer.lower():
            answer += "\n\n⚠️ Ces informations sont à titre éducatif. Consultez un professionnel de santé pour tout conseil médical."

        return answer

    except Exception as e:
        return f"Erreur lors de la génération: {e}"

# Exemple d'utilisation ultra-simple
async def simple_rag_pipeline(query: str, retrieved_docs, mistral_api_key: str = ''):
    """Pipeline RAG simplifié en une fonction"""

    from langchain_mistralai import ChatMistralAI

    # Configuration LLM
    llm = ChatMistralAI(
        model="mistral-medium-latest",
        mistral_api_key=mistral_api_key,
        temperature=0.1
    )

    # Génération
    answer = await generate_simple_response(query, retrieved_docs, llm)

    return answer
```

```python
def setup_rag_components(mistral_api_key: str = ''):
    """Configure les composants RAG (LLM + Reranker)"""

    # Configuration LLM
    llm = ChatMistralAI(
        model="mistral-medium-latest",
        mistral_api_key=mistral_api_key,
        temperature=0.1
    )

    # Configuration Reranker avec le même LLM
    reranker = LLMListwiseRerank.from_llm(llm, top_n=3)

    return llm, reranker
```

```python
async def complete_rag_pipeline(query: str,
                               hybrid_retriever,
                               mistral_api_key: str = '',
                               top_n_rerank: int = 3) -> str:

    print(f"🔍 Recherche pour: '{query}'")

    # 1. RETRIEVAL - Récupération des documents
    documents = hybrid_retriever.get_relevant_documents(query)
    print(f"📄 Documents récupérés: {len(documents)}")

    if not documents:
        return "Aucun document pertinent trouvé pour cette question."

    # 2. SETUP des composants
    llm, reranker = setup_rag_components(mistral_api_key)

    # Mise à jour du top_n du reranker si différent
    if top_n_rerank != 3:
        reranker.top_n = top_n_rerank

    # 3. RERANKING - Amélioration de la pertinence
    print(f"🔄 Reranking vers top-{top_n_rerank}...")
    compressed_docs = reranker.compress_documents(documents, query)
    print(compressed_docs)
    print(f"✅ Documents après reranking: {len(compressed_docs)}")

    # 4. GENERATION - Création de la réponse
    print("🤖 Génération de la réponse...")
    answer = await generate_simple_response(query, compressed_docs, llm)

    return answer , compressed_docs
```

```python
import pandas as pd
import asyncio
from typing import List, Dict
import json
from datetime import datetime

async def create_ragas_test_dataset(
    hybrid_retriever,
    mistral_api_key: str = "",
    test_queries: List[str] = None,
    output_file: str = "ragas_test_dataset.json"
) -> pd.DataFrame:
    """
    Crée un dataset de test pour Ragas en utilisant votre pipeline RAG

    Args:
        hybrid_retriever: Votre retriever configuré
        mistral_api_key: Clé API Mistral
        test_queries: Liste des questions de test (optionnel)
        output_file: Nom du fichier de sortie

    Returns:
        DataFrame avec colonnes: question, contexts, answer, ground_truth
    """

    # Questions de test par défaut pour médecine esthétique
    if test_queries is None:
        test_queries = [
            "Quels sont les effets secondaires du botox ?",
            "Comment se déroule une injection d'acide hyaluronique ?",
            "Quelle est la durée de récupération après un peeling chimique ?",
            "Les injections de botox sont-elles douloureuses ?",
            "Combien coûte un traitement de médecine esthétique ?",
            "Quand peut-on voir les résultats d'un lifting ?",
            "Y a-t-il des contre-indications pour les traitements esthétiques ?",
            "Comment choisir un bon praticien en médecine esthétique ?",
            "Quelle est la différence entre botox et acide hyaluronique ?",
            "Les traitements esthétiques sont-ils remboursés par la sécurité sociale ?"
        ]

    dataset_entries = []

    print(f"🧪 Création du dataset avec {len(test_queries)} questions...")

    for i, query in enumerate(test_queries):
        print(f"\n--- Question {i+1}/{len(test_queries)} ---")
        print(f"📝 {query}")

        try:
            # 1. Récupération des documents
            print("🔍 Récupération des documents...")
            documents = hybrid_retriever.get_relevant_documents(query)

            if not documents:
                print("⚠️ Aucun document trouvé, passage à la question suivante")
                continue

            # Extraire les contextes
            contexts = [
                doc.page_content if hasattr(doc, 'page_content') else str(doc)
                for doc in documents[:5]  # Limiter à 5 contextes max
            ]

            # 2. Setup des composants RAG
            llm, reranker = setup_rag_components(mistral_api_key)
            reranker.top_n = 3

            # 3. Reranking
            print("🔄 Reranking des documents...")
            compressed_docs = reranker.compress_documents(documents, query)

            # Contextes après reranking
            final_contexts = [
                doc.page_content if hasattr(doc, 'page_content') else str(doc)
                for doc in compressed_docs
            ]

            # 4. Génération de la réponse
            print("🤖 Génération de la réponse...")
            answer = await generate_simple_response(query, compressed_docs, llm)

            # 5. Créer l'entrée du dataset
            dataset_entry = {
                "question": query,
                "contexts": final_contexts,
                "answer": answer,
                "ground_truth": answer,  # Pour Ragas, on utilise la réponse comme ground truth
                "metadata": {
                    "original_docs_count": len(documents),
                    "final_docs_count": len(compressed_docs),
                    "query_id": i + 1,
                    "timestamp": datetime.now().isoformat()
                }
            }

            dataset_entries.append(dataset_entry)

            print(f"✅ Question {i+1} traitée avec succès")
            print(f"📄 Contextes: {len(final_contexts)}")
            print(f"📝 Réponse: {answer[:100]}...")

        except Exception as e:
            print(f"❌ Erreur pour la question {i+1}: {e}")
            continue

    # Créer le DataFrame
    df = pd.DataFrame(dataset_entries)

    # Sauvegarder en JSON
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(dataset_entries, f, ensure_ascii=False, indent=2)

    print(f"\n📊 DATASET CRÉÉ:")
    print(f"✅ {len(dataset_entries)} entrées générées")
    print(f"💾 Sauvegardé dans: {output_file}")
    print(f"📋 Colonnes: {list(df.columns)}")

    return df

# Fonction pour tester le dataset créé
async def test_ragas_dataset(dataset_df: pd.DataFrame, sample_size: int = 3):
    """
    Teste le dataset créé avec Ragas sur un échantillon
    """

    print(f"🧪 Test du dataset sur {sample_size} échantillons...")

    # Configuration Ragas
    setup_ragas_with_mistral_fixed("")

    # Prendre un échantillon
    sample = dataset_df.head(sample_size)

    results = []

    for idx, row in sample.iterrows():
        print(f"\n--- Test {idx + 1} ---")
        print(f"Question: {row['question'][:50]}...")

        try:
            # Évaluation avec Ragas
            scores = await score_with_ragas(
                row['question'],
                row['contexts'],
                row['answer']
            )

            results.append({
                "question_id": idx,
                "question": row['question'],
                "scores": scores
            })

            print(f"✅ Scores: {scores}")

        except Exception as e:
            print(f"❌ Erreur: {e}")
            continue

    return results

# Fonction pour charger un dataset existant
def load_ragas_dataset(file_path: str) -> pd.DataFrame:
    """Charge un dataset Ragas depuis un fichier JSON"""

    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    return pd.DataFrame(data)

# Exemple d'utilisation complète
async def create_and_test_dataset(hybrid_retriever):
    """Exemple complet de création et test du dataset"""

    # 1. Créer le dataset
    dataset = await create_ragas_test_dataset(
        hybrid_retriever=hybrid_retriever,
        test_queries=None,  # Utilise les questions par défaut
        output_file="medecine_esthetique_test_dataset.json"
    )

    # 2. Afficher un aperçu
    print("\n📋 APERÇU DU DATASET:")
    print(dataset[['question', 'answer']].head(3))

    # 3. Tester avec Ragas
    test_results = await test_ragas_dataset(dataset, sample_size=2)

    print("\n📊 RÉSULTATS DES TESTS:")
    for result in test_results:
        print(f"Q: {result['question'][:50]}...")
        print(f"Scores: {result['scores']}")
        print()

    return dataset, test_results
```

```python
from langfuse.decorators import observe

@observe(name="evaluate_rag_sample")
def evaluate_rag_sample(sample, faithfulness_score, relevance_score):
    # Simulate processing
    answer = sample["answer"]
    question = sample["question"]

    return {
        "question": question,
        "answer": answer,
        "faithfulness": faithfulness_score,
        "relevance": relevance_score,
        "contexts": sample.get("contexts", [])
    }
```

```python
from langfuse.decorators import observe
from langfuse import get_client

langfuse = get_client()

@observe(name="rag_evaluation")
def run_rag(sample, answer):
    with langfuse.start_as_current_generation(
        name="generate_rag_answer",
        input={"question": sample["question"]}
    ) as gen:
        gen.update(output=answer)
        gen.score(name="faithfulness", value=0.9)
        gen.score(name="relevance", value=0.95)
```
